```python
import builtins

scores  = [78,89,95,96]
total = sum(scores)
avg = total/len(scores)
high = max(scores)
low = builtins.min(scores) # Use builtins.min to avoid shadowing issues

print("Student Scores:",scores)
print("Total Marks:",total)
print("Average Marks:",avg)
print("Highest Mark:",high)
print("lowest Mark:",low)
```
